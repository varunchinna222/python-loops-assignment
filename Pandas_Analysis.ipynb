# Task 1: Load and Inspect the Data

```python
import pandas as pd
import numpy as np

data = {
    'transaction_id': range(1, 21),
    'date': pd.date_range('2024-10-01', periods=20, freq='D'),
    'region': ['North', 'South', 'East', 'West', None, 'North', 'South', None, 'East', 'West',
               'North', 'South', 'East', 'West', 'North', None, 'East', 'West', 'North', 'South'],
    'product_category': ['Electronics', 'Clothing', None, 'Books', 'Electronics', 'Home',
                         'Clothing', 'Books', 'Electronics', None, 'Home', 'Clothing',
                         'Books', 'Electronics', 'Home', 'Clothing', 'Books', 'Electronics',
                         'Home', 'Clothing'],
    'sales_amount': [1200, 450, 890, None, 1500, 670, None, 340, 2100, 780,
                     560, None, 420, 1800, 920, 510, 380, None, 1100, 640],
    'quantity': [2, 5, 3, 1, None, 4, 2, 3, 1, 5,
                 3, 2, None, 1, 4, 3, 2, 1, None, 4],
    'customer_age': [25, 34, None, 45, 29, None, 38, 52, 27, 41,
                     33, None, 48, 26, 35, 42, None, 31, 39, 44],
    'payment_method': ['Credit Card', 'UPI', 'Cash', 'Debit Card', 'Credit Card',
                       'UPI', 'Cash', None, 'Credit Card', 'UPI', 'Debit Card',
                       'Cash', 'Credit Card', None, 'UPI', 'Cash', 'Debit Card',
                       'Credit Card', 'UPI', None]
}

df = pd.DataFrame(data)
df.info()
df.isna().sum()
df
```

# Task 2: Handle Missing Values

```python
#Filling with mode in region and product_category
region_mode = df['region'].mode()[0]
product_category_mode = df['product_category'].mode()[0]
df['region'].value_counts()
df['product_category'].value_counts()
new_df = df.fillna({
    "region": region_mode,
    "product_category": product_category_mode
})
new_df
```

```python
#Filling sales_amount with median
median_sales = df['sales_amount'].median()
new_df = new_df.fillna({
    "sales_amount": median_sales
})
new_df
```

```python
#Filling Quantity using forword fill(ffill)
new_df['quantity'] = new_df['quantity'].ffill()
new_df
```

```python
#Filling Customer age with mean
mean_age = new_df['customer_age'].mean().round(2)
new_df['customer_age'] = new_df['customer_age'].fillna(mean_age)
new_df
```

```python
#Droping rows where payment_method has missing values
cleaned_df = new_df.dropna(subset=['payment_method'])
cleaned_df
```

# Task 3: GroupBy Analysis

```python
#Calculate total sales by region
cleaned_df.groupby('region')['sales_amount'].sum()
```

```python
#Calculate average sales by product_category
cleaned_df.groupby('product_category')['sales_amount'].mean()
```

```python
# Group by both region and product_category, calculate total sales and quantity, then use reset_index() to flatten
cleaned_df.groupby(['region', 'product_category'])[['sales_amount','quantity']].sum().reset_index()
```

```python
# Display top 3 region-product combinations by sales
cleaned_df.groupby(['region', 'product_category'])['sales_amount'].sum().nlargest(3)
```

# Task 4: Custom Aggregation

```python
#Create a function sales_range() that returns max - min of sales
def sales_range(x):
  return x.max() - x.min()

range = cleaned_df.groupby('region')['sales_amount'].agg([sales_range])
print(range)
```

```python
#Use .agg() to calculate multiple metrics by region
cleaned_df.groupby('region').agg({
    'sales_amount':['sum', 'mean', 'max'],
    'quantity': ['sum', 'min']
})
```
